## How To Resample and Interpolate Your Time Series Data With Python

In this tutorial, you will discover how to use Pandas in Python to both increase and decrease the sampling frequency of time series data.

https://machinelearningmastery.com/resample-interpolate-time-series-data-python/

```python
import pandas as pd
from matplotlib import pyplot as plt

series = pd.read_csv('shampoo-sales.csv')
series['Month'] = pd.to_datetime('190' + series['Month'], format='%Y-%m')
series.set_index('Month', inplace=True)
print(series.head())
```

```python
series['Sales'].plot()
plt.show()
```

### Upsample (linear)

```python
upsampled = series.resample('D').interpolate(method='linear')
print(upsampled.head())
```

```python
upsampled.plot()
plt.show()
```

### Upsample (spline)

```python
interpolated = series.resample('D').interpolate(method='spline', order=3)
print(interpolated.head())
```

```python
interpolated.plot()
plt.show()
```

### Downsample

```python
resampled = series.resample('QE').mean()
print(resampled.head())
```

```python
resampled.plot()
plt.show()
```
